```python
import pandas as pd
import numpy as np
```

# Master dataset

```python
df_master = pd.read_pickle("database_2021_11_30.pkl")
df_master.head()
```

```python
x=list(df_master['short_name'].values)
```

```python
#show minimal information
columns = ['rid', 'smiles', 'name_iupac_st', 'name_iupac_so', 'other_names', 'abbreviations','normalized_names']
df_master = df_master[columns]
df_master.head()
```

```python
def check_smiles(name):
    '''
    Return RID and SMILES corresponding to given polymer name
    '''
    for ind in df_master.index:
        names = list(df_master['name_iupac_st'][ind]) + list(df_master['name_iupac_so'][ind]) + list(df_master['other_names'][ind]) + list(df_master['abbreviations'][ind])
        if name in names:
            smiles =df_master['smiles'][ind]
            rid = df_master['rid'][ind]
            break
        else:
            smiles = None
            rid =None
    return rid, smiles
```

# Test dataset

```python
df = pd.DataFrame()
df['name'] = list(['PE', 'PP', 'PMMA', 'polyethylene', 'Polyethylene', 'Poly(ethylene)', 'Polyethylene'])
df
```

```python
#Given polymer names, find smiles suggested by polymer repository

_smiles_strings = []
_rids = []
ctr = 0

print("Searching ...")
print()
for ind in df.index:
    polymer_name = df['name'][ind]
    print(polymer_name)
    try:
        _smiles, _rid = check_smiles(polymer_name)
        if _smiles:
            ctr+=1
            print("found", polymer_name, _smiles, _rid)
    except:
        _smiles, _rid = None, None
        print("not found", polymer_name)
        
    _smiles_strings.append(_smiles)
    _rids.append(_rid)
    print()    
    
df['smiles_suggestion'] = list(_smiles_strings)
df['rid_suggestion'] = list(_rids)
print()
print("Found", ctr, "smiles")
```
