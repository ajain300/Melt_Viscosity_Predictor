```python
data = pd.read_excel('./Data/full_data_03_08_aug_MOD.xlsx', na_values = ['nan','','NaN'])
data.columns = [str(c) for c in data.columns]
Mw_samps, mw_ids = get_Mw_samples(data.copy(), full = True)
```

```python
Mw = np.arange(2, 6, 0.2).reshape(-1,1)
Mw = np.power(10, Mw)
Mw_sc = MinMaxScaler().fit(np.log10(Mw))
error = 0
sample_count = 0
for mid in mw_ids:
    extrap_tests = Mw_test(Mw_samps, mid)
    log_visc_ex = np.array(extrap_tests['exp'][1][~np.isnan(np.power(10,extrap_tests['exp'][1]))]).reshape(-1,)
    Mw_ex = np.array(extrap_tests['exp'][0][~np.isnan((extrap_tests['exp'][1]))]).reshape(-1,)
    #print(Mw_ex, log_visc_ex)
    if Mw_ex.shape[0] > 7 and mid not in [1207, 1230, 1231, 1058, 1204]:
        sample_count += 1
        plt.figure()
        plt.scatter(np.log10(Mw_ex), np.log10(log_visc_ex), c = 'k', label = 'Original Data')
        #try:
        Mw_params = cf.fit_Mw_3(np.log10(Mw_ex), np.log10(log_visc_ex)).tolist()
        visc_true = cf.Mw_func_obj_3(np.log10(Mw),*Mw_params)
        plt.plot(np.log10(Mw), visc_true, '-*', label = r'$f(M)$')
        #except:
        #    print('Mw params not fitting')
      
        #try:
        # Mw1_params, _ = cf.fit_Mw(np.log10(Mw_ex), np.log10(log_visc_ex))
        # plt.plot(np.log10(Mw), cf.Mw_func_obj(np.log10(Mw), *Mw1_params), '-.', label = 'Mw piecewise')
        # #except:
        #    print('Mw1 params not fitting')

      #  try:
        visc_sc = MinMaxScaler().fit(np.log10(log_visc_ex).reshape(-1,1))
        softplus_params = cf.fit_softplus_3(Mw_sc.transform(np.log10(Mw_ex.reshape(-1,1))).reshape(-1,), Mw_sc.transform(np.log10(log_visc_ex.reshape(-1,1))).reshape(-1,)).tolist()
        print('softplus fit', softplus_params)
        visc_est = cf.softplus_3(Mw_sc.transform(np.log10(Mw.reshape(-1,1))).reshape(-1,),*softplus_params)
        
        print('visc_est')
        print(visc_est)
        plt.plot(np.log10((Mw))[~np.isinf(visc_est)], Mw_sc.inverse_transform(visc_est[~np.isinf(visc_est)].reshape(-1,1)).reshape(-1,), '-o', label = r'$\hat{f}(M)$')
      #  except:
     #       print('Softplus params not fitting')
        plt.title(extrap_tests['sample'] + '_' + str(mid))
        plt.legend()
        plt.xlabel('Log Mol Weight')
        plt.ylabel('Log Viscosity')
        plt.show()
        plt.close()
        #print(visc_true)
        #print(visc_est)
        error += mse(visc_true, Mw_sc.inverse_transform(visc_est.reshape(-1,1)).reshape(-1,))
print('Average MSE')
print(error/sample_count)

        
```
