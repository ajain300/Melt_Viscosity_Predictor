```python
from IPython.display import Image
import os
```

```python
no_aug_path = 'Extrapolation_Tests/2022-03-23_full'
aug_path = './Extrapolation_Tests/2022-03-30_full_aug'
os.listdir(no_aug_path)
```

```python
non_aug_ls = {'Mw':{}, 'shear':{}}
for var in ['Mw', 'shear']:
    for model in ['GPR', 'ANN']:
        non_aug_ls[var][model] = os.listdir(f'{no_aug_path}/{var}/{model}/')
aug_ls = {'Mw':{}, 'shear':{}}
for var in ['Mw', 'shear']:
    for model in ['GPR', 'ANN']:
        aug_ls[var][model] = os.listdir(f'{aug_path}/{var}/{model}/')
```

```python
sample = 'cis-1,4-polyisoprene at 25.0 C'
```

```python
for model in ['GPR', 'ANN']:
    display(Image(filename = f'{no_aug_path}/Mw/{model}/' +'/'+str([f for f in non_aug_ls['Mw'][model] if sample in f][0])))
    #plt.imshow(img, aspect='auto')
```

```python
for model in ['GPR', 'ANN']:
    display(Image(filename = f'{aug_path}/Mw/{model}/' +'/'+str([f for f in aug_ls['Mw'][model] if sample in f][0])))
    #plt.imshow(img, aspect='auto')
```
